HOW TO CREATE DATAFRAME

```python
import pandas as pd
```

```python
# First method
list = [['apple','red',11],
['grape','green',22],
['orange','orange',33],
['mango','yellow',44]]

df= pd.DataFrame(list,columns = ["fruits","colour","value"])
print(df)
```

```python
# second metthod
n=['apple','grape','orange','mango']
col = ['red','green','orange','yellow']
val = [11,22,33,44]

dict={'fruits':n,'colour':col,'value':val}
df = pd.DataFrame(dict)
print(df)
```

```python
df.to_csv("fruits.csv",index=False)
```
